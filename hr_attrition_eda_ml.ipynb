```
def plot_cat(attr):
    """Plot count of categories with automatic width adjustment."""

    if attr == 'JobRole':   # Has many categories → wider plot
        sns.catplot(data=df, kind='count',height=3, aspect=3, x=attr)
    else:
        sns.catplot(data=df, kind='count',height=3, aspect=1.7, x=attr)
    plt.title(f"Count Plot of {attr}")
    plt.xlabel(attr)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```
categorical_cols = ['Attrition', 'Gender', 'BusinessTravel', 'Department',
                    'JobRole', 'MaritalStatus', 'EducationField', 'OverTime']

for col in categorical_cols:
    plot_cat(col)
```

```
from sklearn.model_selection import train_test_split
```

```
X=df.drop(columns=['Attrition','BusinessTravel','Over18_encoded','StandardHours','EmployeeCount', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime'])
```

```
Y=df['Attrition']
```

```
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```
from imblearn.over_sampling import SMOTE
```

```
sm=SMOTE(random_state=42)
X_train_sm,Y_train_sm=sm.fit_resample(X_train,Y_train)
```

```
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train_sm)
X_test_scaled=scaler.fit_transform(X_test)
```

```
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report,roc_curve,roc_auc_score
def compare(model):
    clf=model
    clf.fit(X_train_sm,Y_train_sm)
    pred=clf.predict(X_test)

    # Calculating various metrics

    acc.append(accuracy_score(pred,Y_test))
    prec.append(precision_score(pred,Y_test))
    rec.append(recall_score(pred,Y_test))
    auroc.append(roc_auc_score(pred,Y_test))
    classification.append(classification_report(pred,Y_test))
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
acc=[]
prec=[]
rec=[]
auroc=[]
classification=[]
models=[SVC(kernel='rbf'),RandomForestClassifier(),XGBClassifier(),DecisionTreeClassifier()]
model_names=['rbfSVM','RandomForestClassifier','XGBClassifier','DecisionTreeClassifier']

for model in range(len(models)):
    compare(models[model])

d={'Modelling Algo':model_names,'Accuracy':acc,'Precision':prec,'Recall':rec,'Area Under ROC Curve':auroc}

met_df=pd.DataFrame(d)
met_df

#
```

```
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train_scaled,Y_train_sm)
pred=model.predict(X_test_scaled)
print(accuracy_score(Y_test,pred))
```

```
#performance optimization of XGBclassifier
pos = sum(Y_train == 1)
neg = sum(Y_train == 0)
spw = neg / pos
print("scale_pos_weight =", spw)
```

```
from xgboost import XGBClassifier
xgb = XGBClassifier(
    scale_pos_weight=spw,
    max_depth=5,
    learning_rate=0.05,
    n_estimators=800,
    min_child_weight=3,
    subsample=0.9,
    colsample_bytree=0.7,
    gamma=0,                    # allow splits
    reg_alpha=0.1,              # L1 regularization
    reg_lambda=1,               # L2 regularization
    eval_metric='logloss',
    random_state=42,
    tree_method='hist'          # faster, more robust splits
)
xgb.fit(X_train_sm, Y_train_sm)
```

```
import numpy as np
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
y_prob = xgb.predict_proba(X_test)[:, 1]

best_t = 0
best_f1 = 0

for t in np.arange(0.1, 0.9, 0.01):
    y_pred_t = (y_prob >= t).astype(int)
    f1 = f1_score(Y_test, y_pred_t)
    if f1 > best_f1:
        best_f1 = f1
        best_t = t

print("Best threshold:", best_t)

```

```
y_pred = (y_prob >= best_t).astype(int)

print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```

```
import shap

explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X)

shap.summary_plot(shap_values, X, plot_type="bar")

shap_pos=np.where(shap_values>0,shap_values,0)
global_attrition_push=shap_pos.mean(axis=0)

df_global=pd.DataFrame({
    'Feature':X.columns,
    'Mean_SHAP_Value':global_attrition_push
}).sort_values(by="Mean_SHAP_Value",ascending=False)

df_global.to_csv("df_global.csv", index=False)

```

```
import shap
import pandas as pd
import numpy as np

explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X)

shap_df = pd.DataFrame(shap_values, columns=X.columns)


def top3(row):
    top_features = row.nlargest(3).index
    top_values = row[top_features].values
    return pd.Series(list(top_features) + list(top_values))


top3_df = shap_df.apply(top3, axis=1)

top3_df.columns = [
    "TopFeature1", "TopFeature2", "TopFeature3",
    "Value1", "Value2", "Value3"
]
top3_df["EmployeeNumber"] = df["EmployeeNumber"]
top3_df["Attrition_Probability"] = xgb.predict_proba(X)[:,1]


top3_df.to_csv("top3_attrition_factors2.csv", index=False)

top3_df.head()
top3_df
```
